1. Förklara övergripande vad NoSQL är.


Svar: NoSQL är en typ av databashanteringssytem som är lite olikt från en SQLdatabas. NoSQL brukar var bättre lämpade för stora mängder ostrukturerad eller halvstruktruerad data. NoSQL databaser används oftast i moderna applikationer när man arbetar med hög prestanda, skalbarhet och hantering av stora mängder data. Typ big data, molntjänster eller realtidanlyser av data. 

2. Vilken typ av data kan en NoSQL-databas hantera som inte en
relationell databas kan?


Svar: NoSQL databaser kräver inte en fördefinerad struktur över tabeller, kolumer och relationer. Med NoSQL är det där med möjligt att lagra både ostruktuerad data och semistrukturerad data. Till exemple JSON och BSON-dokument. Dokumentdatabaser som MongoDB kan lagra och hantera JSON och BSON format direkt, vilket gör dem mer flexibla än i en relationsdatabas som kräver en fast tabellstruktur.


3. Beskriv kortfattat vad en nyckel/värde-databas är.


Svar: Key-value eller nyckel/värde databas är data kan lagras i nycklar och värden. I en nyckel/värde databas är det endast möjligt att att utföra queries via nycklar och därför behöver dem vara unika. Strukturen i nyckel/värde databas möjliggör lagring av data utan en fördefinerad struktur. Detta medför att databasen kan skalas upp utan att sktrukturen behövs förändras. En fördel med nyckel/värde databaser är när man ska använda sig av snabba queries över enbart nycklar är fyster. Till exempel i en produkt-katalog där man snabbt ska kunna ta ut en produkt. 

4. Beskriv kortfattat vad en dokumentdatabas är.


Svar: Dokumentdatabaser är en typ av NoSQL och i grunden en nyckel/värde databas som lagrar data i ett dokumentformat, oftas en JSON, BSON, XML fil. Varje dokument innehåller en nyckel-värde-par och kan ha en flexibel struktur, vilket gör det möjligt att lagra olika typer va data utan att behöva följa en viss struktur. Typ av program man kan använda till en dokumentdatabas är mongoDB, som är ett verktyg som går att installalera i python. 

5. Beskriv kortfattat vad en bredkolumndatabas är.


Svar: Bredkolumsdatabas är databaser som lagrar data i rader och inte i kolumner. En skillnad är att varje rad kan innehålla olika kolumner. Rader med samma koluner grupperas i kolumner-familjer. Kolumn-familjer lagras oftas på olika diskar. Varje rad behöver inte ha samma kolumner, vilket gör databasen mer anpassningsbar. Apache Cassandra är ett populärt pogram att använda inom bredkolumnsdatabaser. 


6. Beskriv kortfattat vad en grafdatabas är.

Svar: Grafdatabas är en typ av NoSQL-databas som är optimerad för att lagra och hantera relationer mellan data genom att använda noder och kanter istället för tabeller. Den är särskilt kraftfull när data har många samband, som i sociala nätverk, rekommendationssystem och bedrägeridetektion. Neo4J är en populärt program att göra grafdatabaser genom att till exempel analysera sociala nätverk.

```python
# Query för att få ut vilka produktersom behövs beställa hem
query = {
    "$expr": {
        "$gt": [
            "$ReorderLevel",
            {"$sum": ["$UnitsInStock", "$UnitsOnOrder"] }
        ]
    }
}

products_to_order = collection.find(query)

for product in products_to_order:
    print(product)
```

```python
# Att man ser 
pipeline = [
    {
        "$project": {
            "_id": 1,
            "CompanyName": "$SupplierDetail.CompanyName",
            "Phone": "$SupplierDetail.Phone",
            "ContactName": "$SupplierDetail.ContactName"
        }
    }
]

suppliers = list(collection.aggregate(pipeline))
if suppliers:
    collection.insert_many(suppliers)

    print("SupplierDetails har extraherats och sparats i 'suppliers' samlingen.")
```

```python
pipeline = [
    {
        "$lookup": {
            "from": "SuppliersDetails",  
            "localField": "SupplierDetails.supplierID",  
            "foreignField": "supplierID",  
            "as": "supplier_Details"
        }
    },
    {
        "$addFields": {
            "supplierdetail": { "$arrayElemAt": ["$supplier_Details", 0] } 
        }
    },
    {
        "$project": { "supplier_info": 0 } 
    }
]

orders_with_supplier = list(collection.aggregate(pipeline))
```

```python
with open("movies.json", "r") as f:
    movies = json.load(f)
```

```python



with open("movies.json", "r", encoding="utf-8") as file:
    movies_data = json.load(file)

# Skapa unika ID:n för actors, directors och genres
actor_dict = {}
director_dict = {}
genre_dict = {}

# Filinnehåll
movies_list = []
actors_movies_list = []
actors_list = []
directors_list = []
genres_movies_list = []
genres_list = []

# Generera data för CSV-filer
for movie in movies_data:
    movie_id = movie["movie_id"]
    
    
    movies_list.append([
        movie_id, movie["movie_title"].strip(), movie["title_year"], 
        movie["director_name"], movie["duration"], movie["gross"], 
        movie["num_voted_users"], movie["language"], movie["country"], 
        movie["content_rating"], movie["budget"], movie["title_year"], 
        movie["imdb_score"], movie["aspect_ratio"]
    ])
    
    
    if movie["director_name"] not in director_dict.keys():
        director_id = len(director_dict) + 1
        director_dict[movie["director_name"]] = director_id
        directors_list.append([director_id, movie["director_name"]])

    # Actors.csv & Actors_movies.csv
    for actor_name in movie["actors"]:
        if actor_name not in actor_dict.keys():
            actor_id = len(actor_dict) + 1
            actor_dict[actor_name] = actor_id
            actors_list.append([actor_id, actor_name])
        actors_movies_list.append([actor_dict[actor_name], movie_id])

    # Genres.csv & Genres_movies.csv
    for genre in movie["genres"]:
        if genre not in genre_dict:
            genre_id = len(genre_dict) + 1
            genre_dict[genre] = genre_id
            genres_list.append([genre_id, genre])
        genres_movies_list.append([genre_dict[genre], movie_id])

# Funktion för att skriva CSV-filer
def write_csv(filename, data, headers):
    with open(filename, "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        writer.writerows(data)

# Skapa CSV-filer
write_csv("movies.csv", movies_list, ["movie_id", "movie_title", "year", "director", "duration", "gross", "num_voted_users", "language", "country", "content_rating", "budget", "title_year", "imdb_score", "aspect_ratio"])
write_csv("actors_movies.csv", actors_movies_list, ["actor_id", "movie_id"])
write_csv("actors.csv", actors_list, ["actor_id", "actor_name"])
write_csv("directors.csv", directors_list, ["director_id", "director_name"])
write_csv("genres_movies.csv", genres_movies_list, ["genre_id", "movie_id"])
write_csv("genres.csv", genres_list, ["genre_id", "genre"])
```
